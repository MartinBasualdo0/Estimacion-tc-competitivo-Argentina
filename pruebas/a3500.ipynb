```python
import pandas as pd
import plotly.graph_objects as go
```

```python
cotizaciones_original = pd.read_excel("../data/cotizaciones 1997.xlsx").set_index("Período")
cotizaciones_original = cotizaciones_original.resample("ME").mean()
```

```python
com3500 = pd.read_excel("https://www.bcra.gob.ar/Pdfs/PublicacionesEstadisticas/com3500.xls", header=3)
com3500 = com3500.dropna(axis=1, thresh=1)\
    .dropna()\
    .drop("Tipo de Cambio Nominal Promedio Mensual", axis=1)\
    .set_index("Fecha")\
    .resample("ME").first()
```

```python
cotizaciones_original
```

```python
df = com3500.join(cotizaciones_original)
df["dif%"] = df["Tipo de Cambio de Referencia - en Pesos - por Dólar"] / df["Estados Unidos"] - 1
fig = go.Figure()
# fig.add_trace(go.Scatter(y = df["Tipo de Cambio de Referencia - en Pesos - por Dólar"], x= df.index, name = "A3500" ))
# fig.add_trace(go.Scatter(y = df["Estados Unidos"], x= df.index, name = "Cotizaciones" ))
fig.add_trace(go.Scatter(y = df["dif%"], x= df.index, name = "Diferencia%",
                         hovertemplate="Date: %{x}<br>" +
                  "Difference: %{y:.5%}<extra></extra>",))
fig.update_layout(template=None, title_text = "Diferencia entre los tipos de cambio 'AL3500' y el otro publicado")
fig.update_yaxes(tickformat = ".0%")
```
